```python
import random
import string

print("~~Strong Password Generator~~")

uzunluk = int(input("Şifre uzunluğu giriniz (4-15): "))

if uzunluk < 4 or uzunluk > 15:
    print("Lütfen 4 ile 15 arasında bir sayı giriniz.")
else:
    sifre = [
        random.choice(string.ascii_lowercase),
        random.choice(string.ascii_uppercase),
        random.choice(string.digits),
        random.choice(string.punctuation)
    ]

    karakterler = string.ascii_letters + string.digits + string.punctuation

    while len(sifre) < uzunluk:
        sifre.append(random.choice(karakterler))
    
    random.shuffle(sifre)

    print("\nOluşturulan Güçlü Şifre:")
    print("".join(sifre))

```
